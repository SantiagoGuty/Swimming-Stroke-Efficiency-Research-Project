```python

import matplotlib.lines as mlines

# Step 3: Create a DataFrame
data = pd.DataFrame({'Name': name, 'Strokes': strokes, 'Time': time, 'Division': division})

# Calculate the mean time for each unique stroke count
mean_time_by_strokes = data.groupby('Strokes')['Time'].mean().reset_index()

# Step 4: Use Seaborn to create the plot with hue
sns.set(style="whitegrid")  # Set seaborn style
plt.figure(figsize=(10, 6))  # Set the figure size

# Create the scatter plot using Seaborn's scatterplot with hue
sns.scatterplot(x='Strokes', y='Time', data=data, hue='Division', palette='flare', s=100, zorder=2)

# Plot the regression line
reg_line = sns.regplot(x='Strokes', y='Time', data=data, scatter=False, line_kws={'alpha': 0.5}, color='#4e88c7')

# Plot the line of the mean time for each stroke count
sns.lineplot(x='Strokes', y='Time', data=mean_time_by_strokes, color='#ff8080', sort=False, zorder=1)

# Create custom legend handles
custom_handles = [mlines.Line2D([], [], color='#4e88c7', label='Regression Line'),
                  mlines.Line2D([], [], color='#ff8080', label='Mean Time')]

# Add title and labels
plt.title('\nRelationship Between Stroke Count and Times in 100 backstroke \n', fontsize=18)
plt.xlabel('Stroke Count')
plt.ylabel('Time')

# Show the legend
plt.legend(handles=custom_handles, title='Division', loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(False)

# Show the plot
plt.show()
```
